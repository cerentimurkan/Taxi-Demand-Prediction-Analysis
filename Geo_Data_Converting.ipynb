```python
import pandas as pd
import numpy as np

import geopandas as gpd
from shapely.geometry import Point, LineString, Polygon
```

```python
df_geo=gpd.read_file(r"C:\Users\asd\geoproject\shapefiles\taxi_zones.shp")
df_geo = df_geo.drop(['OBJECTID','Shape_Leng','Shape_Area'], axis=1)
df_geo.rename(columns={"LocationID":"PULocationID"},inplace=True)
df_geo = df_geo.to_crs(epsg=4326)
df_geo.head()
```

```python
coord_list = [(x,y) for x,y in zip(df_geo["geometry"].centroid.x,df_geo["geometry"].centroid.y)]
coord_list
```

```python
df_coor = pd.DataFrame(data=coord_list)
df_geo["Longitude"] = df_coor[0]
df_geo["Latitude"] = df_coor[1]
df_geo = pd.DataFrame(df_geo)
df_geo.drop(columns="geometry",inplace=True)
df_geo.head()
```

```python
df_geo.to_csv("geo_coor.csv")
```
